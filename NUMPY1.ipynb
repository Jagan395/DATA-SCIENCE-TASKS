```python
#Write a function to create a 2d array with 1 on the border and 0 inside. Take 2-D array shape as (a,b) as parameter to function.
def create2D(a,b):
    a1=np.ones((a,b))
    a1[1:a-1,1:b-1]=0
    return a1
ar=create2D(6,6)
print(ar)
```

```python
#Softmax function
arr=np.array([86.03331084 ,37.7285648  ,48.64908087, 87.16563062 ,38.40852563 ,37.20006318])
def softmax(arr):
    if arr.ndim !=1 :
        raise ValueError("give me 1D array")
    res=np.exp(arr)/np.sum((np.exp(arr)))
    return res
ans=softmax(arr)
ans
```
