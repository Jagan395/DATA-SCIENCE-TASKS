```python
#plot the tanh function
import matplotlib.pyplot as plt
x=np.arange(1,11)
y=(np.exp(x)-np.exp(-(x)))/(np.exp(x)+np.exp(-(x)))
plt.plot(x,y)
plt.show()

```
